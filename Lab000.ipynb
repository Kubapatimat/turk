Użytkowanie interfejsów:
Interfejsem w systemie Cisco IOS nazywamy encję, która wprowadza lub
wyprowadza dane z urządzenia (jego systemu). Interfejsy mogą być
odpowiednikami urządzeń fizycznych (np. utożsamione z gniazdami w obudowie
fizycznej urządzenia, Np. GigabitEthernet), lub logiczne (ustanowione
programowo – np. Tunnel, Dialer, Loopback, BVI itp.). Posiadają też różne
funkcjonalności (np. mogą funkcjonować w ramach pakietu IP lub nie). Znaczną
cześć ustawień konfiguracyjnych prowadzi się wokół interfejsów. Sporą część ich
typów ustanawia się także dynamicznie.
Sprawdzenie możliwych do ustanowienia w systemie typów interfejsów IP:

`Switch#show ip interface ?`

````python
t("show ip interface ?")
````

Jedną z częściej używanych komend jest komenda generująca raport o
ustanowionych w systemie interfejsach IP:
`Switch#show ip interface brie`
```
Interface IP-Address OK? Method Status Protocol
GigabitEthernet0/1 unassigned YES manual down down
GigabitEthernet0/2 unassigned YES manual down down
GigabitEthernet0/3 unassigned YES manual down down
GigabitEthernet0/4 unassigned YES manual down down
GigabitEthernet0/5 unassigned YES manual down down
...
```
Zweryfikuj, czy lista interfejsów odpowiada gniazdom w obudowie przełącznika.
W raporcie szczególną na uwagę zasługują ostatnie dwie kolumny tego raportu.
W sieciach komputerowych urządzenie sprawne, aktywne, włączone określamy
słowem „up”, wyłączone, niesprawne, usunięte – słowem „down”. Pierwsza z
kolumn (status) określa fizyczny stan medium (elektrycznego, radiowego,
optycznego itp.). Przypadek szczególny: stan Administratively down – wyłączone
w konfiguracji przez administratora.

Druga kolumna (protocol) określa kompatybilność (ustawień) urządzenia
komunikującego się z nami przez medium lub naszą gotowość do nawiązania
takiej komunikacji (interfejs od strony protokołu czyli zbioru zasad komunikowania
się jest ustawiony prawidłowo).

Podłącz kabel Ethernet prowadzący z Twojej stacji PC do przełącznika
(wybierając dowolny jego port Ethernet). Obserwuj komunikaty pojawiające się na
ekranie. Następnie ponów komendę:

`Switch#show ip interface brief`

sprawdzając statusy portów.

Informacja praktyczna: W Cisco nie ma konieczności typowania pełnych słów
kluczowych komend czy ich parametrów. Wystarczy podawanie ich unikatowych
przedrostków. Przykładowo: jeśli nie istnieje inna komenda rozpoczynająca się od
liter „sh” jak „show” wystarczy wytypowanie „sh”. W konsekwencji – poprzednią
komendę można zapisać następująco:

`Switch#sh ip int br`

Dodatkowo posiadamy (jak to często bywa w systemach CLI) funkcję auto-
uzupełnienia komendy i jej parametru dostępną pod klawiszem <TAB>:

`Switch#sho<TAB>`

Warto z niej korzystać w celu łatwego nauczanie się komend oraz weryfikacji
poprawności ich wprowadzenia (funkcja działa tylko wtedy gdy wcześniejsze
człony komendy są zapisane poprawnie).

````python
t("show ip interface brief")
````

````python
t("sh ip int br")
````

W większości systemów zarządzających urządzeniami sieciowymi funkcjonalność
(i interfejs CLI) zapewniająca utrzymanie przełącznika jest wyraźnie
odseparowana od umożliwiającej konfigurowanie jego zachowania w sieci. Dzieje
się tak głównie ze względów bezpieczeństwa. Mamy zatem osobny tryb
konfiguracji (nazwany config, edit etc.), w którym dokonujemy zmian. Sama
diagnostyka prowadzona jest z trybu exec.

Aby przejść do trybu konfiguracji należy posłużyć się komenda configure z
parametrem określającym źródło konfiguracji (pamięć, sieć, plik i w końcu
terminal przy którym siedzimy):

`Switch#configure terminal`

`Switch(config)#`

Zwróć uwagę na zmianę w karetce systemu operacyjnego. Znajdujemy się
obecnie w trybie konfiguracji i dostępne komendy przeznaczone są do
prowadzenia w niej zmian.

Uwaga: W Cisco IOS pula dostępnych komend zmienia się każdorazowo, gdy
zmieniamy tryb pracy lub typujemy bardziej szczegółowo przedmiot konfiguracji
(inaczej, niż bywa w systemach operacyjnych dla stacji roboczych czy serwerów).
Sprawdź dostępną pulę zagadnień związanych z konfigurowaniem przełącznika:

`Switch(config)# ?`

rozwijając listę spacją.

Uwaga: we wszystkich dalszych pakietach laboratoryjnych każdy przykład
komendy jest poprzedzany zapisem określającym stan karetki, przy jakim
komendę należy wydać. Należy go uważnie obserwować.

````python
t("configure terminal")
````

````python
t("?")
````

Zagadnienia będące przedmiotem konfiguracji można typować podając ich
nazwę. W tym momencie karetka przemieszcza się do kolejnego trybu,
dotyczącego danego zagadnienia i pula komend zmienia się (na dotyczącą

gdzie `GigabitEthernet0/5` to nazwa interfejsu pobrana z wcześniej rozważanej
listy (należy podać nazwę właściwego interfejsu). Zauważ, że karetka znów się
zmieniła. Sprawdź pulę komend dostępną dla wybranego interfejsu:

`Switch(config-if)# ?`

````python
t("interface FastEthernet0/0")
````

Dokonamy teraz najprostszej konfiguracji interfejsu. Będzie to jego wyłączenie (w
konfiguracji). W tym celu wydaj komendę:

`Switch(config-if)#shutdown`

i obserwuj zachowanie interfejsu (sygnalizatory LED w obudowie przełącznika,
komunikaty). Zweryfikuj (komendą `show ip interface brief`) stan interfesju.

Uwaga – komendy tej nie możemy wydać bezpośrednio w trybie konfiguracji.
Konieczne byłoby wyjście z tego trybu (komenda exit lub sekwencja klawiszy
CTRL-z), wydanie tej komendy, a następnie powrót w celu kontynuowania
konfigurowania urządzenia.
```
Switch(config-if)#exit
Switch(config)#exit
Switch#show ip interface brief
Switch#configure terminal
Switch(config)#interface GigabitEthernet0/5
Switch(config-if)#
```
Tą skomplikowaną procedurę można uprościć poprzez zastosowanie przedrostka
„do” przed komendą trybu exec, będąc w trybie konfiguracji. Uprawnia nas to do
wydania wtedy komendy trybu exec:

`Switch(config-if)#do show ip interface brief`

````python
t("shutdown")
````

````python
t("exit",
  "exit",
  "show ip interface brief",
  "configure terminal",
  "interface FastEthernet0/0")
````

````python
t("do show ip interface brief")
````

Pliki konfiguracji przełącznika

Wszelkie zmiany konfiguracyjne dokonane w systemie IOS przełącznika Cisco
przez administratora trzymane są w jednym spójnym zasobie – pliku konfiguracji.
Format (składnia) przechowywanej tam konfiguracji jest identyczna jak komend,
które taką konfigurację stanowią (dotyczy to także komend przejść). Dlatego
konfiguracji można używać jako komend (wklejając ją do terminala celem
wykonania), a komend jako konfiguracji (zapisując w pliku). Aby zobaczyć
zawartość pliku konfiguracji, która aktualnie działa, należy posłużyć się komendą:

`Switch#show running-config`

````python
t("do show running-config")
````

Znak „!” oznacza komentarz (instrukcję pustą).
Prezentowana powyżej konfiguracja jest typową konfiguracją „fabryczną”,
przeznaczoną do uzupełniania i powodującą, że podstawowe usługi przełącznika
od razu działają. Warto ją przestudiować, aby móc w przyszłości wykryć zmiany
dokonane przez późniejsze czynności. Zmiany takie mogą zakłócać przebieg
doświadczeń, dlatego należy je wykryć/usunąć.
Gdy chcemy usunąć wpis z konfiguracji – stosujemy tą komendę ponownie, ale z
przedrostkiem „no”:

`Switch(config)#Interface GigabitEthernet0/5`

`Switch(config-if)#no shutdown`

````python
t("no shutdown")
````

Istnieje jeszcze drugi plik konfiguracji o szczególnym znaczeniu – zapisana w
pamięci trwałej konfiguracja startowa – startup-config. Możliwe jest kopiowanie
plików w obydwu kierunkach:

`Switch#copy running-config startup-config`

Zapisanie konfiguracji startowej spowoduje jej użycie w czasie uruchamiania
przełącznika (już jako running-config). Możliwe jest skasowanie pamięci trwałej:

`Switch#write erase`

aby przywrócić konfigurację fabryczną przy następnym uruchomieniu
przełącznika

````python
t("exit", "exit")
````

````python
t("copy running-config startup-config", "\r")
````

````python
t("write erase\r")
````
